```python
import numpy as np  
import pandas as pd
```

```python
names = [j for j in list('Input {}'.format(i) for i in range(1,91))]
names.append('class')
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/libras/movement_libras.data', names = names)
data.head()
```

```python
X = data.drop(['class'],axis=1)
y = data['class']
display(X.describe())
```

```python
class PC:
  def PCA(X):
      mean_X = X - np.mean(X , axis = 0)
      mat_cov = np.cov(mean_X , rowvar = False)
      eigen_values , eigen_vectors = np.linalg.eigh(mat_cov)
      sorted_index = np.argsort(eigen_values)[::-1]
      sorted_eigenvalue = eigen_values[sorted_index]
      sorted_eigenvectors = eigen_vectors[:,sorted_index]
      return sorted_eigenvalue, sorted_eigenvectors
```

```python
from sklearn.preprocessing import StandardScaler

#Standardise the data to fit into our PCA algorith
X = StandardScaler().fit_transform(X)
print(X)
```

```python
eigenval, eigenvec = PC.PCA(X)
print("Eigen Values")
print(eigenval, "\n")
print(pd.DataFrame(eigenval).describe())
print("\n\n\n")
print("Eigen Vectors")
print(eigenvec, "\n")
print(pd.DataFrame(eigenvec).describe())
```
