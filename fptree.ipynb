```python
def filetodic(filename):
  data = pd.read_csv(filename)
  npdata = np.array(data)
  # npdata = npdata[:1000]
  mp = dict()
  cnt=1
  for i in npdata:
    mp[cnt] = i[0].split()
    cnt+=1
  return mp

def itemlistgen(mapofdata):
  itemset = set([])
  for key in mapofdata:
    tempset = set(mapofdata[key])
    itemset = tempset.union(itemset)
  return list(itemset)

def get_item_with_sup(mapofdata):
  itemap = dict()
  for key in mapofdata:
    # itemset = frozenset({mapofdata[key]})
    for itm in mapofdata[key]:
      if frozenset({itm}) in itemap.keys():
        itemap[frozenset({itm})]+=1
      else:
        itemap[frozenset({itm})]=1
  return itemap

def get_item_with_min_support(itemap, min_sup):
  item_greater_than_min_support = {}
  for key in itemap:
    if itemap[key]>min_sup:
      item_greater_than_min_support[key]=itemap[key]
  return item_greater_than_min_support

def support(items_counts, mapofdata):
    support = dict()
    for i in items_counts:
        support[i] = items_counts[i]/len(mapofdata)
    return support


#This Function calculates Support value of given set
def computeSupport(item_with_sup, mapofdata):
  sup=0
  itemset = frozenset(item_with_sup)
  for i in mapofdata:
    setofcur = frozenset(mapofdata[i])
    if itemset.issubset(setofcur):
      sup+=1
  return sup


def freq(item_list_k, mapofdata):
  item_count = dict()
  for item in item_list_k:
    temp_i = set(item)
    for j in mapofdata.items():
      if temp_i.issubset(set(j[1])):
        if item in item_count:
          item_count[item]+=1
        else:
          item_count[item]=1
  return item_count

def intialset (dataset):
    temp_mp = {}
    for trans in dataset:
        temp_mp[frozenset(trans)] = 1
    return temp_m
```
